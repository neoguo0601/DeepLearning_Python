```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# Load an color image in grayscale
img = cv2.imread('sample.jpg',cv2.IMREAD_UNCHANGED)
```

```python
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('saverImage.png',img)
```

```python
img = cv2.imread('sample.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
# Create a black image
img = np.zeros((512,512,3), np.uint8)

# Draw a diagonal blue line with thickness of 5 px
image = cv2.line(img,(0,0),(511,511),(255,0,0),5)
#需要把BGR转换成RGB
plt.imshow(image[:,:,::-1], interpolation = 'bicubic')
plt.show()
```

```python
img = np.zeros((512,512,3), np.uint8)
image = cv2.rectangle(img,(384,3),(510,128),(0,255,0),3)
plt.imshow(image[:,:,::-1], interpolation = 'bicubic')
plt.show()
```

```python
img = np.zeros((512,512,3), np.uint8)
image = cv2.circle(img,(447,63), 63, (0,0,255), -1)
plt.imshow(image[:,:,::-1], interpolation = 'bicubic')
plt.show()
```

```python
img = np.zeros((512,512,3), np.uint8)
image = cv2.ellipse(img,(256,256),(100,50),0,0,180,255,-1)
plt.imshow(image[:,:,::-1], interpolation = 'bicubic')
plt.show()
```

```python
img = np.zeros((512,512,3), np.uint8)
pts = np.array([[20,15],[60,150],[100,150],[100,80]], np.int32)
pts = pts.reshape((-1,1,2))
cv2.polylines(img,[pts],True,(0,255,255), 3)
plt.imshow(img[:,:,::-1], interpolation = 'bicubic')
plt.show()
```

```python
img = np.zeros((512,512,3), np.uint8)
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,'OpenCV',(10,500), font, 4,(255,255,255),2,cv2.LINE_AA)
plt.imshow(img[:,:,::-1], interpolation = 'bicubic')
plt.show()
```
